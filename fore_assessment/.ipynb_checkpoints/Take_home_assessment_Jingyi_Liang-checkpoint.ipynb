# Data Processing

## Feature Engineering

```python
#Employee tenure
df_all['tenure'] = 0
for i in df_all['tenure'].index:
    if pd.isnull(df_all.iloc[i, 4])==True:
        df_all.iloc[i, -1] = (pd.to_datetime('today') - df_all.iloc[i, 3]).days/365.25
    else:
        df_all.iloc[i, -1] = (df_all.iloc[i, 4] - df_all.iloc[i, 3]).days/365.25
df_all.head()
```

```python
#Create Labels for Voluntary Turnover
df_all['is_voluntary'] = np.where(df_all['Termination Type']=='Voluntary', True, False)
```

```python
#Create Lag Features
df_all.sort_values(by = ['Employee ID', 'Performance Month'], ascending = True, inplace = True)
df_all['Schedule Adherence Lag'] = df_all.groupby('Employee ID')['Schedule Adherence'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Schedule Variability Lag'] = df_all.groupby('Employee ID')['Schedule Variability'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Error Rate Lag'] = df_all.groupby('Employee ID')['Error Rate'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Performance Score Lag'] = df_all.groupby('Employee ID')['Performance Score'].transform(lambda x: x.rolling(3, closed = 'left').mean())
```

```python
#Create Percentile in Department Features
df_all['Schedule Adherence Percentile'] = df_all.groupby('Department')['Schedule Adherence'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Variability Percentile'] = df_all.groupby('Department')['Schedule Variability'].transform(lambda x: x.rank(pct=True))
df_all['Error Rate Percentile'] = df_all.groupby('Department')['Error Rate'].transform(lambda x: x.rank(pct=True))
df_all['Performance Score Percentile'] = df_all.groupby('Department')['Performance Score'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Adherence Lag Percentile'] = df_all.groupby('Department')['Schedule Adherence Lag'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Variability Lag Percentile'] = df_all.groupby('Department')['Schedule Variability Lag'].transform(lambda x: x.rank(pct=True))
df_all['Error Rate Lag Percentile'] = df_all.groupby('Department')['Error Rate Lag'].transform(lambda x: x.rank(pct=True))
df_all['Performance Score Lag Percentile'] = df_all.groupby('Department')['Performance Score Lag'].transform(lambda x: x.rank(pct=True))
```

# Model Input Preparation

```python
#Prepare target variable format
df_model = df_all.copy()
df_model.drop('Termination Type', axis = 1, inplace = True)
df_model['is_voluntary'] = np.where(df_model['is_voluntary']==True, 1, 0)
```

```python
#Handling Inapplicable Rows & Checking types
df_model.drop(['Name', 'Start Date', 'End Date'], axis = 1, inplace = True)
##Creating dummy variables
df_model = pd.get_dummies(df_model, prefix=['Department', 'State', 'City'], columns=['Department', 'State', 'City'])
```

```python
#Checking columns types
df_model.dtypes
```

```python
#Filling missing values
##Locating columns with missing value
df_model.isnull().sum()
```

```python
#After exploring we can find out that 'Population', 'Median House Price' and 'Average Commute Time' missing values are those rows with state AZ
#Since this is missing not at random, the best way is actually to find out external data to fill in (importing population resource)
#However, since external data is prohibited here, I'll fill with -1 to remind the model that this state's data are different from others
df_model['Population'].fillna(-1, inplace=True)
df_model['Median House Price'].fillna(-1, inplace=True)
df_model['Average Commute Time'].fillna(-1, inplace=True)
#filling the Lags columns, similar to above, those are also not missing at random
df_model['Schedule Adherence Lag'].fillna(-1, inplace=True)
df_model['Schedule Variability Lag'].fillna(-1, inplace=True)
df_model['Error Rate Lag'].fillna(-1, inplace=True)
df_model['Performance Score Lag'].fillna(-1, inplace=True)
df_model['Schedule Adherence Lag Percentile'].fillna(-1, inplace=True)
df_model['Schedule Variability Lag Percentile'].fillna(-1, inplace=True)
df_model['Error Rate Lag Percentile'].fillna(-1, inplace=True)
df_model['Performance Score Lag Percentile'].fillna(-1, inplace=True)
```

```python
#Dataset Partitioning
##Sort data by year month
df_model.sort_values(by = ['Year', 'Performance Month'], inplace = True)
##Split out training set
x_cols = list(df_model.columns.drop(['is_voluntary']))
y_col = 'is_voluntary'
train, test1 = train_test_split(df_model, test_size = 0.4, random_state = 1, shuffle = False)
x_train = train.loc[:,x_cols]
y_train = train[y_col]
##Split test and validation set
test, val = train_test_split(test1, test_size = 0.15, random_state = 1, shuffle = False)
x_test = test.loc[:,x_cols]
y_test = test[y_col]
x_val = val.loc[:,x_cols]
y_val = val[y_col]
```

```python
##Department Slices
def plot_department_turnover(department):
    df_turnover = df_all[df_all['Department'] == department].copy()
    df_turnover['Location'] = df_turnover['City']+', '+df_turnover['State']
    df_turnover['Turnover Rate'] = df_turnover.groupby(['Year', 'Performance Month','Location'])[['is_voluntary']].transform('sum')/df_turnover.groupby(['Year', 'Performance Month','Location'])[['is_voluntary']].transform('count')
    df_turnover = df_turnover[['Location', 'Year', 'Performance Month', 'Turnover Rate']]
    df_turnover.drop_duplicates(inplace = True)
    df_turnover['Time'] = pd.to_datetime(df_turnover['Year'].astype(str) + '-' + df_turnover['Performance Month'].astype(str) + '-01')
    ###Calculating all employees turnover rate in the slice
    df_turnover2 = df_all[df_all['Department'] == department].copy()
    df_turnover2['Turnover Rate']= df_turnover2.groupby(['Year', 'Performance Month'])[['is_voluntary']].transform('sum')/df_turnover2.groupby(['Year', 'Performance Month'])[['is_voluntary']].transform('count')
    df_turnover2 = df_turnover2[['Year', 'Performance Month', 'Turnover Rate']]
    df_turnover2.drop_duplicates(inplace = True)
    df_turnover2['Time'] = pd.to_datetime(df_turnover2['Year'].astype(str) + '-' + df_turnover2['Performance Month'].astype(str) + '-01')
    df_turnover2['Location'] = 'All Employees'
    df_turnover = df_turnover.append(df_turnover2)
    ###Plotting
    fig, ax = plt.subplots(figsize = (10, 8))
    palette = {c:'red' if c=='All Employees' else default_palette for c in df_turnover['Location'].unique()}
    sns.lineplot(data = df_turnover, x = 'Time', y = 'Turnover Rate', hue = 'Location', palette=palette)
    plt.title('Monthly Trunover Rate - '+department)
    plt.show()
###Plot for all departments
for i in df_all['Department'].unique():
    plot_department_turnover(i)
```

## Logistic Regression

```python
#Output is a binary variable, so start with a logistic regression model
log_model = linear_model.LogisticRegression(max_iter=6000)
log_model.fit(x_train,y_train)
predictions_test = log_model.predict(x_test)
actuals_test = np.array(y_test)
predictions_val = log_model.predict(x_val)
actuals_val = np.array(y_val)
logistic_acc_test = log_model.score(x_test,actuals_test)
logistic_acc_val = log_model.score(x_val,actuals_val)
y_test_pred = log_model.predict(x_test)
y_val_pred = log_model.predict(x_val)
##mean squared error
mse_test = mean_squared_error(y_test, y_test_pred, squared=True)
mse_val = mean_squared_error(y_val, y_val_pred, squared=True)
##root mean squared error
rms_test = mean_squared_error(y_test, y_test_pred, squared=False)
rms_val = mean_squared_error(y_val, y_val_pred, squared=False)
df_log_result = pd.DataFrame({'Accuracy': [logistic_acc_test, logistic_acc_val], 'MSE': [mse_test, mse_val], 'RMSE':[rms_test, rms_val]}, index = ['Test Set', 'Validation Set'])
df_log_result
```

```python
##Define a function for calculating confusion matrix
def get_confusion_matrix(y_test, predictions):
    tn, fp, fn, tp = confusion_matrix(np.array(y_test), predictions).ravel()
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    tpr = tp/(tp+fn)
    fpr = fp/(fp+tn)
    f_score = 2*precision*tpr/(precision+tpr)
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    misclassification = (fp + fn)/(tp+fp+tn+fn)
    print("Precision:\t\t\t%1.2f"%(precision))
    print("Recall/TPR:\t\t\t%1.2f"%(recall))
    print("False Positive Rate:\t\t%1.2f"%fpr)
    print("f-score:\t\t\t%1.2f tradeoff between precision and recall"%(f_score))
    print("Accuracy:\t\t\t%1.2f how well the model has classified"%(accuracy))
    print("Misclassification:\t\t%1.2f ratio of mistakes"%(misclassification))
##Confusion matrix for test set
get_confusion_matrix(y_test, predictions_test)
```

```python
##Confusion matrix for validation set
get_confusion_matrix(y_val, predictions_val)
##Accuracies for logistic model are high, and even higher for validation set, showing a better prediction power for more recent data
```
