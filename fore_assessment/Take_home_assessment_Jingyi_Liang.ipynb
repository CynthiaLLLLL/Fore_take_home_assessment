## Data Integration

```python
df_all = df_performance.merge(df_employee, how = 'left', left_on = 'Employee ID', right_on = 'ID')
df_all = df_all.merge(df_city, how = 'left', left_on = 'City', right_on = 'City')[['Employee ID', 'Performance Month', 'Name', 'Start Date', 'End Date', 'Termination Type', 'Department', 'State_x', 'City', 'Schedule Adherence', 'Schedule Variability', 'Error Rate', 'Performance Score', 'Elevation', 'Latitude', 'Longitude']]
df_all.rename(columns={'State_x':'State'}, inplace=True)
df_all['Year'] = df_all['Performance Month'].dt.year
df_all = df_all.merge(df_state, how = 'left', left_on = ['State', 'Year'], right_on = ['State', 'Year'])
```

# Data Processing

## Feature Engineering

```python
## Employee tenure
df_all['tenure'] = 0
for i in df_all['tenure'].index:
    if pd.isnull(df_all.iloc[i, 4])==True:
        df_all.iloc[i, -1] = (pd.to_datetime('today') - df_all.iloc[i, 3]).days/365.25
    else:
        df_all.iloc[i, -1] = (df_all.iloc[i, 4] - df_all.iloc[i, 3]).days/365.25
df_all.head()
```

```python
## Create Labels for Voluntary Turnover
df_all['is_voluntary'] = np.where(df_all['Termination Type']=='Voluntary', True, False)
```

```python
##Create Lag Features
df_all.sort_values(by = ['Employee ID', 'Performance Month'], ascending = True, inplace = True)
df_all['Schedule Adherence Lag'] = df_all.groupby('Employee ID')['Schedule Adherence'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Schedule Variability Lag'] = df_all.groupby('Employee ID')['Schedule Variability'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Error Rate Lag'] = df_all.groupby('Employee ID')['Error Rate'].transform(lambda x: x.rolling(3, closed = 'left').mean())
df_all['Performance Score Lag'] = df_all.groupby('Employee ID')['Performance Score'].transform(lambda x: x.rolling(3, closed = 'left').mean())
```

```python
##Create Percentile in Department Features
df_all['Schedule Adherence Percentile'] = df_all.groupby('Department')['Schedule Adherence'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Variability Percentile'] = df_all.groupby('Department')['Schedule Variability'].transform(lambda x: x.rank(pct=True))
df_all['Error Rate Percentile'] = df_all.groupby('Department')['Error Rate'].transform(lambda x: x.rank(pct=True))
df_all['Performance Score Percentile'] = df_all.groupby('Department')['Performance Score'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Adherence Lag Percentile'] = df_all.groupby('Department')['Schedule Adherence Lag'].transform(lambda x: x.rank(pct=True))
df_all['Schedule Variability Lag Percentile'] = df_all.groupby('Department')['Schedule Variability Lag'].transform(lambda x: x.rank(pct=True))
df_all['Error Rate Lag Percentile'] = df_all.groupby('Department')['Error Rate Lag'].transform(lambda x: x.rank(pct=True))
df_all['Performance Score Lag Percentile'] = df_all.groupby('Department')['Performance Score Lag'].transform(lambda x: x.rank(pct=True))
```
